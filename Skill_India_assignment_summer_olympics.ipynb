```python
import pandas as pd
df = pd.read_csv('summer_olympics.csv')
df.head()
```

Q 1. In how many cities Summer Olympics is held so far?

```python
print('There are ',len(df['City'].unique()),' number of cities in which olympics is held so far.')
print('which are ',df['City'].unique())
```

Q 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
data2=[]
for sport in df['Sport'].unique():
    data2.append([sport,len(df[ (df['Sport']==sport) & (df['Medal']=='Gold') ]) ])
pd.DataFrame(data2,columns=['Sports','Gold medal']).sort_values(by='Gold medal',ascending=False).head().plot(x='Sports',y='Gold medal',kind='bar',figsize=(10,5))
```

Q 3. Which sport is having most number of medals so far? (Top 5)

```python
data3=[]
for sport in df['Sport'].unique():
    data3.append([sport,len(df[df['Sport']==sport])])
pd.DataFrame(data3,columns=['Sports','Medals']).sort_values(by='Medals',ascending=False).head().plot(x='Sports',y='Medals',kind='bar',figsize=(10,5))
```

Q 4. Which player has won most number of medals? (Top 5)

```python
data4=[]
for Athlete in df['Athlete'].unique():
    data4.append([Athlete,len(df[df['Athlete']==Athlete])])
pd.DataFrame(data4,columns=['Athlete','Medals']).sort_values(by='Medals',ascending=False).head().plot(x='Athlete',y='Medals',kind='bar',figsize=(10,5))
```

Q 5. Which player has won most number Gold Medals of medals? (Top 5)

```python
data5=[]
for Athlete in df['Athlete'].unique():
    data5.append([Athlete,len(df[ (df['Athlete']==Athlete) & (df['Medal']=='Gold') ])])
pd.DataFrame(data5,columns=['Athlete','Gold Medals']).sort_values(by='Gold Medals',ascending=False).head().plot(x='Athlete',y='Gold Medals',kind='bar',figsize=(10,5))
```

Q 6. In which year India won first Gold Medal in Summer Olympics?

```python
#method 1
yo = df.sort_values(by = ['Year'])
for i in range(len(yo)):
    if(yo.loc[i]['Country']=='IND' and yo.loc[i]['Medal']=='Gold'):
        print('India won first olympics Gold in year ',yo.loc[i]['Year'])
        break
```

```python
#method 2
y = df[(df['Country']=='IND')&(df['Medal']=='Gold')].sort_values(by='Year')
print('India won first olympics Gold in year ',y['Year'].tolist()[0])
```

Q 7. Which event is most popular in terms on number of players? (Top 5)

```python
#calculating events based on number of occurence in dataframe
data7=[]
for Event in df['Event'].unique():
    data7.append([Event,len(df[df['Event']==Event])])
pd.DataFrame(data7,columns=['Event','frequency']).sort_values(by='frequency',ascending=False).head().plot(x='Event',y='frequency',kind='bar',figsize=(10,5))
```

Q 8. Which sport is having most female Gold Medalists? (Top 5)

```python
data8=[]
for Sport in df['Sport'].unique():
    data8.append([Sport,len(df[(df['Sport']==Sport)&(df['Gender']=='Women')&(df['Medal']=='Gold')])])
pd.DataFrame(data8,columns=['Sport','Women']).sort_values(by='Women',ascending=False).head().plot(x='Sport',y='Women',kind='bar',figsize=(10,5))
```
